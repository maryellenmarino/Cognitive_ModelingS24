```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def monte_carlo_approx_pi(num_points):
    points = np.random.uniform(-1, 1, size=(2, num_points))
    def lies_in_circle(point):
        return point[0]**2+point[1]**2<1
    in_circle = lies_in_circle(points)
    num_in_circle = in_circle.sum()
    return 4 * (num_in_circle/num_points)
```

```python
vals_to_test = [10,100,1000,10000,100000,1000000,10000000,100000000]
```

```python
approximations = np.array([monte_carlo_approx_pi(val) for val in vals_to_test])
approximations
```

```python
errors = abs(approximations-np.pi)
```

```python
errors
```

```python
plt.title("Number of Values vs. Pi Error") 
plt.xlabel("Number of Monte Carlo Values") 
plt.ylabel("Error of Pi")  
plt.xscale("log")
plt.yscale("log")
plt.plot(vals_to_test, errors, color ="maroon") 
```
