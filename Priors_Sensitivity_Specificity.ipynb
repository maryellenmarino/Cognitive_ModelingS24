```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
priors = np.linspace(0,1,101)
```

```python
def posteriorFromPrior(prior):
    sensitivity = 0.95
    specificity = 0.90
    probTruePositive = prior*sensitivity
    probFalsePositive = (1-prior)*(1-specificity)
    probPosTest = probTruePositive+probFalsePositive
    normalizedTruePositive = probTruePositive/probPosTest;
    return normalizedTruePositive
```

```python
posteriorValues = posteriorFromPrior(priors)
```

```python
posteriorValues
```

```python
plt.title("Prior vs. Posterior") 
plt.xlabel("Percent of Population with Disease") 
plt.ylabel("Probability of having Disease Given Positive Test")  
plt.plot(priors, posteriorValues, color ="maroon") 
```

```python
sensitivities = np.linspace(0,1,101)
```

```python
def posteriorFromSensitivity(sensitivity):
    prior = 0.01
    specificity = 0.90
    probTruePositive = prior*sensitivity
    probFalsePositive = (1-prior)*(1-specificity)
    probPosTest = probTruePositive+probFalsePositive
    normalizedTruePositive = probTruePositive/probPosTest;
    return normalizedTruePositive
```

```python
posteriorValues1 = posteriorFromSensitivity(sensitivities)
```

```python
posteriorValues1
```

```python
plt.title("Sensitivity vs. Posterior") 
plt.xlabel("Probability of a True Positive") 
plt.ylabel("Probability of having Disease Given Positive Test") 
plt.plot(sensitivities, posteriorValues1, color ="maroon") 
```
