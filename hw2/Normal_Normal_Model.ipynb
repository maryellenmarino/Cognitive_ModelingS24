```python
sigma_squared = 2
data = np.random.normal(loc=72, scale=2, size=12)
y=np.average(data)
data
```

```python
y
```

He then delved into his database of crimes and found that the average height of criminals is 70 inches with a standard deviation of 3 inches. This would be his prior

```python
mu_zero = 70
sigma_zero = 3
observation = y
```

```python
prior = stats.norm(loc=mu_zero, scale=sigma_zero)
```

He then simulated his prior and posterior to see what heights of students he should focus on most. 

```python
x = np.linspace(60,80,10000)
prior_density = prior.pdf(x)
```

```python
loc_posterior = ((sigma_zero**2)/(sigma_squared+(sigma_zero**2)))*observation+(sigma_squared/(sigma_squared+(sigma_zero**2)))*mu_zero
scale_posterior = 1/( ( 1/( sigma_zero**2 ) )+( 1/sigma_squared ) )
posterior = stats.norm(loc=loc_posterior, scale=scale_posterior)
posterior_density = posterior.pdf(x)
```

```python
f, ax = plt.subplots(1, 1, figsize = (10,6))
ax.plot(x, prior_density, color = 'maroon', label = 'Prior height of criminals')
ax.plot(x, posterior_density, color = 'gray', label = 'Posterior height of criminals')
ax.legend()
sns.despine(ax=ax)
```
