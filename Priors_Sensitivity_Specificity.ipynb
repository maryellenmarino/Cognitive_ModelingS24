```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
priors = np.linspace(0.01,1,100)
```

```python
def posteriorFromPrior(prior):
    sensitivity = 0.95
    specificity = 0.90
    probTruePositive = prior*sensitivity
    probFalsePositive = (1-prior)*(1-specificity)
    probPosTest = probTruePositive+probFalsePositive
    normalizedTruePositive = probTruePositive/probPosTest;
    return normalizedTruePositive
```

```python
posteriorValues = posteriorFromPrior(priors)
```

```python
posteriorValues
```

```python
plt.title("Prior vs. Posterior") 
plt.xlabel("Percent of Population with Disease") 
plt.ylabel("Probability of having Disease Given Positive Test")  
plt.plot(priors, posteriorValues, color ="maroon") 
```

```python
specificities = np.linspace(0.01,1,100)
```

```python
def posteriorFromSpecificity(specificity):
    prior = 0.01
    sensitivity = 0.95
    probTruePositive = prior*sensitivity
    probFalsePositive = (1-prior)*(1-specificity)
    probPosTest = probTruePositive+probFalsePositive
    normalizedTruePositive = probTruePositive/probPosTest;
    return normalizedTruePositive
```

```python
posteriorValues2
```

```python
plt.title("Specificity vs. Posterior") 
plt.xlabel("Probability of a True Negative") 
plt.ylabel("Probability of having Disease Given Positive Test") 
plt.plot(specificities, posteriorValues2, color ="maroon") 
```

```python
def posteriorFromPriorAndSensitivity(prior, sensitivity):
    specificity = 0.9
    probTruePositive = prior*sensitivity
    probFalsePositive = (1-prior)*(1-specificity)
    probPosTest = probTruePositive+probFalsePositive
    normalizedTruePositive = probTruePositive/probPosTest;
    return normalizedTruePositive
```

```python
X, Y = np.meshgrid(priors, sensitivities)
X,Y
```

```python
posteriors = posteriorFromPriorAndSensitivity(X,Y)
```

```python
fig = plt.figure(figsize =(6, 6))
ax = plt.axes(projection ='3d')
my_cmap = plt.get_cmap('hot')
surf = ax.plot_surface(priors, sensitivities, posteriors, cmap = my_cmap, edgecolor ='none')
fig.colorbar(surf, ax = ax, shrink = 0.5, aspect = )
plt.show()
```
