```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
mu_zero = 0
sigma_zero = 1
sigma_squared = 1
observation = 1
```

```python
prior = stats.norm(loc=mu_zero, scale=sigma_zero)
```

```python
x = np.linspace(-3,3,10000)
prior_density = prior.pdf(x)
```

```python
loc_posterior = ((sigma_zero**2)/(sigma_squared+(sigma_zero**2)))*observation+(sigma_squared/(sigma_squared+(sigma_zero**2)))*mu_zero
scale_posterior = 1/( ( 1/( sigma_zero**2 ) )+( 1/sigma_squared ) )
posterior = stats.norm(loc=loc_posterior, scale=scale_posterior)
posterior_density = posterior.pdf(x)
```

```python
f, ax = plt.subplots(1, 1, figsize = (10,6))
ax.plot(x, prior_density, color = 'maroon', label = 'Prior')
ax.plot(x, posterior_density, color = 'gray', label = 'Posterior')
ax.legend()
sns.despine(ax=ax)
```
