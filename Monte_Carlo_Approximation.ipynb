```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
def monteCarloApproxPi(numPoints):
    points = np.random.uniform(-1, 1, size=(2, numPoints))
    def liesInCircle(point):
        return point[0]**2+point[1]**2<1
    inCircle = liesInCircle(points)
    numInCircle = inCircle.sum()
    return 4 * (numInCircle/numPoints)
```

```python
valsToTest = [10,100,1000,10000,100000,1000000,10000000,100000000]
```

```python
approximations = np.array([monteCarloApproxPi(val) for val in valsToTest])
approximations
```

```python
errors = abs(approximations-np.pi)
```

```python
errors
```

```python
plt.title("Number of Values vs. Pi Error") 
plt.xlabel("Number of Monte Carlo Values") 
plt.ylabel("Error of Pi")  
plt.xscale("log")
plt.plot(valsToTest, errors, color ="maroon") 
```
